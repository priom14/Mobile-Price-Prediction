## Models 

#### Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
X = df.drop(columns = ['price_range'], axis = 1)
Y = df['price_range']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print('number of rows in train data is', X_train.shape)
print('number of rows in test data is', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
DT_model = DecisionTreeClassifier()
DT_model.fit(X_train, y_train)
```

```python
y_pred = DT_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('\33[46mModel Accuracy Score: {0:0.5f}'. format(accuracy_score(y_test, y_pred)))
print('training set score: {:.5f}'.format(DT_model.score(X_train, y_train)))
print('test set score: {:.5f}'.format(DT_model.score(X_test, y_test)))
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10, shuffle=False) #cross-validation
score = cross_val_score(DT_model, X_train, y_train, cv=kf, scoring='accuracy')
DT_model_cv_score = score.mean()
DT_model_cv_stdev = np.std(score)
print('cross validation accuracy scores are:\n {}'.format(score))
```

```python
accuracy = ['CV Accuracy']
DT_A = pd.DataFrame({'CV MEAM':DT_model_cv_score,'STD':DT_model_cv_stdev},index=accuracy)
DT_A
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.2)
ConfusionMatrixDisplay.from_estimator(DT_model, X_test, y_test, colorbar=False, cmap='GnBu')
plt.title('Confusion Matrix of Base Decision Tree')
plt.grid(False)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
feature_names = ['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']
target_names = ["0", "1", "2", "3"]
fig = plt.figure(figsize = (30, 15))
Plot = plot_tree (DT_model, feature_names = feature_names, class_names = target_names, filled = True)
```

```python
importances = DT_model.feature_importances_ # feature importances
indices = np.argsort(importances) #sort
plt.figure(figsize=(10, 6))
plt.barh(range(len(indices)), importances[indices], color='#52796f', align='center')
plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
plt.xlabel('Relative Importance')
for i, v in enumerate(importances[indices]):
    plt.text(v, i, str(round(v, 3)), color='black', fontsize= 8)
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
def metrics_calculator(y_test, y_pred, model_name):
    result = pd.DataFrame(data=[accuracy_score(y_test, y_pred),
                                precision_score(y_test, y_pred, average='macro'),
                                recall_score(y_test, y_pred, average='macro'),
                                f1_score(y_test, y_pred, average='macro')],
                          index=['Accuracy','Precision','Recall','F1-score'],
                          columns = [model_name])
    return result
```

```python
BaseDT = metrics_calculator(y_test, y_pred, 'Base Decision Tree')
BaseDT
```

### Random Forest Classifier

```python
RF_model = RandomForestClassifier()
RF_model.fit(X_train, y_train) # fit model
```

```python
y_pred = RF_model.predict(X_test)
```

```python
print('\33[46mModel Accuracy Score: {0:0.5f}'. format(accuracy_score(y_test, y_pred)))
print('training set score: {:.5f}'.format(RF_model.score(X_train, y_train)))
print('test set score: {:.5f}'.format(RF_model.score(X_test, y_test)))
shapes = {
    'X_train': X_train.shape[0],
    'y_train': y_train.shape[0],
    'X_test': X_test.shape[0],
    'y_test': y_test.shape[0]}
plt.figure(figsize=(16, 6))
plt.bar(shapes.keys(), shapes.values(),color="darkseagreen")
plt.ylabel('Count')
plt.title('Distribution of Training & Validation Sets')
plt.show()
```

```python
kf = KFold(n_splits=10, shuffle=False) #cross-validation
score = cross_val_score(RF_model, X_train, y_train, cv=kf, scoring='accuracy')
RF_model_cv_score = score.mean()
RF_model_cv_stdev = np.std(score)
print('cross validation accuracy scores are:\n {}'.format(score))
```

```python
Accuracy = ['CV Accuracy']
RFA = pd.DataFrame({'CV Mean':RF_model_cv_score,'STD':RF_model_cv_stdev},index=Accuracy)
RFA
```
